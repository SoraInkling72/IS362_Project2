## Analyses

### What dessert(s) have more than one occurrence in the DataFrame? Or rather, what desserts come from multiple countries/regions? 
These multiple occurrences indicate that these desserts are shared among different cultures.

The 'value_counts()' function counts the number of occurrences based on its series. If there are mulitple of the same value in a series, this function will identify how many times it occurs.

```python
shared = desserts['Dessert Name'].value_counts()
multicultural = pd.DataFrame(shared)
multicultural.columns = ['Countries/Regions Present']
multicultural[multicultural['Countries/Regions Present'] > 1]
```

### Answer:

**Polvorónes**, **Borma**, and **Pavlova** are the desserts that have more than one occurrence in the DataFrame. Thus, these desserts are shared among different cultures.

## Dataset 2 - Mobile Device Management (MDM), by Ziwei Song

```python
mdm = pd.read_csv('dummy_devices.csv')
mdm
```

To clean up and tidy the data, I would first have to remove all of the rows with devices that have no owners or emails registered. Devices that have no owners thus do not meet the MDM compliance requirements and should not have  access to the organization's data and resources. To do this, I would have to use the "dropna.()" function to remove all of those rows. Also, I would have to specify that it only drops the rows that have "NaN" values for the "Owner" and "Username" columns only. This is because one person, Simon English, has an iPhone 14 Pro Max registered to the system, but it is not recorded whether or not the device is MDM compliant, as there is no value for it present. Judging by the lack of a customized name for the device, it can be assumed that the device is either new or underwent a factory reset; though it is more likely the former as it has the newest version of iOS installed.

```python
clean_mdm = mdm.dropna(subset=['Owner', 'Username'])
clean_mdm
```

### Who has an MDM compliant PC (Laptop or Desktop) that runs on the latest Windows OS?

```python
pc_sort = clean_mdm[clean_mdm['Operating Systems'] == 'Windows']
pc_sort.sort_values('Version', ascending=False)
```

To find the answer, I had to first filter the data that ran on the Windows OS, and then sort out the values of the version of the OS, but the values had to descend so that the latest version would appear on the first few rows. Thus, it would show the full details of the registered device: the owners, the device model, and whether or not it is MDM compliant.

### Answer:

**Salvador Nash** has an MDM compliant desktop that runs on the latest Windows OS, Version 11.0.26100, on an HP Z2 Mini G9.

### Who has mulitple devices registered into organization's system?

In order to determine this, I would have to count all of the occurrences of names that appear in the dataset. As the same name registered for multiple devices would indicate that the person has multiple ways to access the system, working across devices. On the dataset, this would be shown if a value for "Owner" occurs more than once.

```python
multi_device = clean_mdm['Owner'].value_counts()
devices = pd.DataFrame(multi_device)
devices.columns = ['Devices Owned']
devices[devices['Devices Owned'] > 1]
```

### Answer:

## Dataset 3 - Software Company Employee Data

```python
employee_info = pd.read_csv('Software_Company_Employees.csv')
employee_info
```

### Who is the youngest Software Engineer in the entire corporation? Where are they stationed?

```python
software_engineer = employee_info[employee_info['Position'] == 'Software Engineer']
young_se = software_engineer.sort_values('Age')
young_se
```

To be able to find the answer for this, I would first have to filter out the software engineers in the company from the "Position" series, creating an object ("software_engineer") out of the expression to do so. Afterwards, I would create another object ("young_se") to sort out the employees based on their age with ascending values. The first row would have the lowest value present, that being the software enginer who is the youngest amongst them, along with other information about them.

### Answer:

The youngest software engineer in the entire corporation is **Sonya Frost**, stationed in **Edinburgh, Scotland**.

### Who was employed at the corporation the longest? Or rather, who has the earliest start date?

The person who has the earliest start date is the one who has been employed at the company for the longest. Thus, to find the answer, the data would be parsed and sorted base on the series, "Start date." However, before using the 'sort_values()' function, the date values would have to be converted to datetime datatype objects with 'pd.to_datetime(). Afterwards, then the values in the dataset can be sorted, with the earliest employee being in the first row in the dataframe.

```python
employee_info['Start date'] = pd.to_datetime(employee_info['Start date'])

employee_info.sort_values('Start date').head()
```

### Answer:
